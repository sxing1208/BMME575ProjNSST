```python
import os
import requests

# List of image URLs
image_urls = []

def get_url(number):
    numstr = str(number).zfill(3)
    return f"https://www.med.harvard.edu/AANLIB/cases/case29/mr1/{numstr}.gif"

for i in range(5,24):
    image_urls.append(get_url(i))

# Directory to save images
save_dir = './alz/mri'
os.makedirs(save_dir, exist_ok=True)  # Create directory if it doesn't exist
# Function to download a single GIF
def download_gif(url, save_path):
    try:
        response = requests.get(url, stream=True)
        if response.status_code == 200:
            with open(save_path, 'wb') as file:
                for chunk in response.iter_content(1024):
                    file.write(chunk)
            print(f"Downloaded {save_path}")
        else:
            print(f"Failed to download {url} - Status Code: {response.status_code}")
    except Exception as e:
        print(f"An error occurred while downloading {url}: {e}")

# Batch download GIFs
for idx, url in enumerate(image_urls):
    gif_name = f"{idx + 1}.gif"
    save_path = os.path.join(save_dir, gif_name)
    download_gif(url, save_path)
```

```python
# MRI data
# Categorize data sets, save as 60% train, 20% validation, 20% test
!pip install opencv-python
!pip install tensorflow
import cv2
import os
import numpy as np
from tensorflow.keras.preprocessing.image import img_to_array, load_img
from sklearn.model_selection import train_test_split

label_types = ['alz', 'brchgen', 'hyp', 'motor', 'nrm']

for label in label_types:
    cd = os.curdir
    m_path = os.path.join(cd, label, 'mri')
    working_list = os.listdir(m_path)
    ret_list = []
    
   
    for i in range(len(working_list)):
        src = os.path.join(m_path, str(i+1) + '.gif')
        img = load_img(src, color_mode='rgb')
        im = img_to_array(img)
        ret = cv2.resize(im, (256, 256))  
        ret_list.append(ret)
    
    ret_array = np.array(ret_list)
    ret_array = ret_array.astype('float32') / 255.0 # image normalization
    
    # 60% train, 20% validation, 20% test
    X_train, X_temp = train_test_split(ret_array, test_size=0.4, random_state=42)
    X_val, X_test = train_test_split(X_temp, test_size=0.5, random_state=42)

    # transpose arrays to (256, 256, 3, n)
    X_train = np.transpose(X_train, (1, 2, 3, 0))
    X_val = np.transpose(X_val, (1, 2, 3, 0))
    X_test = np.transpose(X_test, (1, 2, 3, 0))


    np.save(f'{label}_mri_train.npy', X_train)
    np.save(f'{label}_mri_val.npy', X_val)
    np.save(f'{label}_mri_test.npy', X_test)
    
    print(f"{label} MRI - Train set shape:", X_train.shape)
    print(f"{label} MRI - Validation set shape:", X_val.shape)
    print(f"{label} MRI - Test set shape:", X_test.shape)
```

```python
# PET Data
import cv2
import os
import numpy as np
from tensorflow.keras.preprocessing.image import img_to_array, load_img
from sklearn.model_selection import train_test_split

label_types = ['alz', 'brchgen', 'hyp', 'motor', 'nrm']

for label in label_types:
    cd = os.curdir
    m_path = os.path.join(cd, label, 'pet')
    working_list = os.listdir(m_path)
    ret_list = []
    
    for i in range(len(working_list)):
        src = os.path.join(m_path, str(i+1) + '.gif')
        img = load_img(src, color_mode='rgb')
        im = img_to_array(img)
        ret = cv2.resize(im, (256, 256))  
        ret_list.append(ret)
    
    ret_array = np.array(ret_list)
    ret_array = ret_array.astype('float32') / 255.0 
    
  
    X_train, X_temp = train_test_split(ret_array, test_size=0.4, random_state=42)
    X_val, X_test = train_test_split(X_temp, test_size=0.5, random_state=42)

    X_train = np.transpose(X_train, (1, 2, 3, 0))
    X_val = np.transpose(X_val, (1, 2, 3, 0))
    X_test = np.transpose(X_test, (1, 2, 3, 0))


    np.save(f'{label}_pet_train.npy', X_train)
    np.save(f'{label}_pet_val.npy', X_val)
    np.save(f'{label}_pet_test.npy', X_test)
    
    print(f"{label} PET - Train set shape:", X_train.shape)
    print(f"{label} PET - Validation set shape:", X_val.shape)
    print(f"{label} PET - Test set shape:", X_test.shape)
```
